## InsertionSort - Ordenação por Inserção

1. Nesse algoritmo será eleito o segundo número do vetor para iniciar as comparações.
2. Os elementos masi à esquerda do número eleito estão sempre ordenados de forma crescente ou decrescente.
3. Um laço de comparação será executado do segundo elemento ao último.
4. Essa repetição se realizará na quantidade de vezes igual ao número de elementosd do vetor menos um (n-1).

## InsertSort em Python

```python
n = int(input("Entre com o número de elementos de X: "))

#Criando a lista de 5 elementos a partir da entrada do usuário
X = [int(input("Digite o {0}{1}: ".format(i+1, chr(176)))) for i in range(n)]
print("O vetor de entrada é:  ")
print(X)
#Ordenação de forma crescente, a partir do número de elementos em X

#Laço com a quantidade de elementos do vetor
for i in range(1, n):
    eleito = X[i]
    j = i - 1
    while j >=0 and X[j] > eleito:
        X[j+1] = X[j]
        j = j - 1
    X[j+1] = eleito
        
print("O vetor ordenado é: ")
print(X)
```

## MergerSort - Algoritmo de ordenação por intercalação

1. O vertor é dividido em vetores com a metade do tamanho original por meio de um procedimento recursivo
2. Essa divisão ocorre até que o vetor fique com apenas um elemento e estes sejam ordenados e intercalados.
3. Usaremos a técnica da divisão e conquista, que envolve três passos:
    - Dividir o problema em um certo número de subproblemas.
    - Conquistar os subproblemas solucionando-os recursivamente. Se os tamanhos dos subproblemas são suficientemente pequenos, então solucionar os subproblemas de forma simples
    - Combinar as soluções dos subproblemas na solução do problema original


1. No MergeSort, usa-se a técnica de dividir e conquistar da seguinte forma:
    - Dividir: dividir a sequência de n elementos a serem ordenados em duas subsequencias de $n/2$ elementos cada.
    - Conquistar: ordenar as duas subsequências recursivamente utilizando ordenação por intercalação
    - Combinar: intercalar as duas subsequências ordenadas para produziar a solução.

## MergeSort em Python

```python
def merge(X, inicio, fim):
    if inicio < fim:
        meio = int((inicio + fim) / 2)
        merge(X, inicio, meio)
        merge(X, meio + 1, fim)
        intercalar(X, inicio, meio, fim)

        
        
def intercalar(X, inicio, meio, fim):
    aux = [] 
    i, j = inicio, meio + 1
    poslivre = inicio

    while poslivre < fim + 1:
        if i > meio:
            aux.append(X[j])
            j += 1
        elif j > fim:
            aux.append(X[i])
            i += 1
        elif X[j] < X[i]:
            aux.append(X[j])
            j += 1
        else:
            aux.append(X[i])
            i += 1
            
        poslivre = poslivre + 1

    poslivre = 0
    for i in range(inicio, fim+1):
        X[i] = aux[poslivre]
        poslivre = poslivre + 1

    

n = int(input("Entre com o número de elementos de X: "))

#Criando a lista de 5 elementos a partir da entrada do usuário
X = [int(input("Digite o {0}{1}: ".format(i+1, chr(176)))) for i in range(n)]
print("O vetor de entrada é:  ")
print(X)
merge(X, 0, n-1)
print("O vetor ordenado é: ")
print(X)



    
```

1. Primeiro precisamos saber o tempo do algoritmo intercalar.
2. Observe que não temos um laço dentro de outro laço nesse novo algoritmo.
3. Como o intercalar irá fazer a intercalação entre doi vetores de tamanho $m_{1}$ e $m_{2}$, então o tempo é $n = m_{1} + m_{2}$
4. Note que fazemos a chamada recursiva da função `merge` duas vezes, sendo que em cada chamada passa apenas a metade do vetor, ou seja $n/2$ para cada chamada.
5. Sabendo então que a função intercalar será executada $n$ vezes, a função que representa esse algorítmo será:
    - $T(n) = T(\frac{n}{2}) + T(\frac{n}{2}) + n$
    - $T(n) = 2T(\frac{n}{2}) + n$


####  Notação $\Theta$ e método master

1. Notação $\Theta$ é utilizada para limite assintótico firme sobre uma função.
2. Nesse caso a função $f(n)$ é $\Theta(g(n))$ se existem constantes positivas $c_{1}$, $c_{2}$ e $n_{0}$ tais que $0 \le c_{1} g(n) \le f(n) \le c_{2}g(n)$, para todo $n \ge n_{0}$
3. Exemplo: $f(n) = 1/2n^{2}-3n$, $f(n)$ é $\Theta(n^2)$, pois c_{1} n^2 \le f(n) \le c_{2}n^2$, quando $c_{1} = 1/8$, $c_{2} = 1/2$ e $n_{0}=8$.

1. O método master apresenta um teorema para resolver quase todas as recorrências $T(n)$ que possuam a forma $aT(\frac{n}{b})+f(n)$ para $a,b>1$
2. Para isso fazemos:
    - Se $f(n) = O(n^{log_{b}{a-\varepsilon}})$, para alguma constante $\varepsilon > 0$, então $T(n) = \Theta(n^{\log_{b}{a}})$
    - Se  $f(n) = \Theta(n^{\log_{b}{a}})$ então $T(n) = \Theta(n^{\log_{b}{a}}\log(n))$
    - Se $f(n) = \Omega(n^{log_{b}{a-\varepsilon}})$ para alguma constante $\varepsilon > 0$, se $a f(\frac{n}{b}) \le c f(n)$ para alguma constante $c < 1$ e para todo $n$ suficientemente grande, então $T(n) = \Theta(f(n))$

1. Para o Algoritmo MergeSort podemos ver que:
    - $T(n) = 2T(\frac{n}{2}) + n$
    - Assim, $a=2$, $b=2$ e $f(n)=n$
    - Em linhas gerais, podemos ver que $f(n) = \Theta(n^{\log_{b}{a}})$
    - Logo: $n = \Theta(n^{\log_{2}{1}})$
    - Que leva: $n = \Theta(n^{1})$
    - Pelo método master, podemos concluir que $T(n) = \Theta(n\log(n))$


## Heapsort
1. Baseado na estrutura HEAP, que nada é que um vetor (X) que pode ser visto como uma árvore binária completa, em que cada nó possui no máximo 2 filhos.
2. Cada vértice da árvore corresponde a um elemento do vetor.
3. A árvore é completamente preenchida exceto no último nível.
4. Cada nível é sempre preenchido da direita para a esquerda.
5. Num HEAP, para todo vértice i diferente da raiz, a seguinte propriedade dever ser válida: X[Pai(i)] $\ge$ X[i]

1. Dado um índice i do vetor: 
    - $Pai(i) = i / 2)$
    - $Filho_{esquerdo}(i) = 2* i$
    - $Filho_{direito}(i) = 2 * i + 1$
2. Como o HEAP é um vetor, o maior elemento ficará na rais, então o primeiro elemento pode ser trocado com o último, ocupando a posição correta dentro do vetor.
3. Após feita a troca, desconsidera-se esse último elemento.
4. Esse processo continua até a árvore permanecer com um único elemento

```python
def transformar_heap(arr, n, i):
    #Inicializa a Raíz com o maior valor
    largest = i  # Initialize largest as root
    # left = 2*i + 1
    l = 2 * i + 1  
    # right = 2*i + 2
    r = 2 * i + 2  
    
    #Verifica se o filho esquerdo da raiz existe e se é a maior raiz
    if l < n and arr[i] < arr[l]:
        largest = l
        
    #Verifica se o filho direito da raiz existe e se ela 
    # é maior que a raiz

    if r < n and arr[largest] < arr[r]:
        largest = r
 
    #Troca a raíz, se necessário
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]  # troca 
        # Heapify the root.
        transformar_heap(arr, n, largest)
        
# Função principal de ordenação do vetor dado
def heapSort(arr):
    n = len(arr)
 
    # Construídno o maxheap
    for i in range(n, -1, -1):
        transformar_heap(arr, n, i)
 
    # Um por um extraímos os elementos
    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]   # troca
        transformar_heap(arr, i, 0)
        
n = int(input("Entre com o número de elementos de X: "))

#Criando a lista de 5 elementos a partir da entrada do usuário
X = [int(input("Digite o {0}{1}: ".format(i+1, chr(176)))) for i in range(n)]
print("O vetor de entrada é:  ")
print(X)


heapSort(X)
print("O vetor ordenado é: ")
print(X)
```

### Complexidade do HeapSort

1. A função `heapSort` faz a chamada do método `transformar_heap` duas vezes, em dois laços distintos.
2. O `heapSort` chama o `transformar_heap` $n$ vezes.
3. O método `transformar_heap` é aplicado sempre a um nó da árvore, que representa um elemento do vetor, e afunda esse nó até que a propriedade HEAP seja válida.
4. O pior caso ocorre quando quando o procedimento é aplicado sobre o primeiro elemento e este deve afundar até alcançar uma folha.
5. Logo, o número de trocas corresponde à altura da árvore, que é $\log n$.
6. Dessa forma, vemos que a complexidade desse algorítmo é $O(n\log{(n)})$
