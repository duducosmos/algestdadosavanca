## Expansão Telescópica

Considere a relação de Recorrência:

$ T(n) = T(\frac{n}{2})+1$

Para resolver essa equação, aplicamos $T(\frac{n}{2})$ na fórmula de $T(n)$ e obtemos:

$T(\frac{n}{2}) = T(\frac{n}{4})+1$

$ T(n) = T(\frac{n}{2})+1$
$ T(n) = (T(\frac{n}{4})+1)+1$
$ T(n) = ((T(\frac{n}{8})+1)+1)+1$
...
$ T(n) = (...((((T(\frac{n}{2^k})+1+1)+1)+...+1)$

Vamos assumir que $n$ é uma potência de 2, logo $n=2^k$. Também consideramos que $T(n/2^k)=T(1) = 1$

Logo:

$ T(n) = (...((((T(\frac{n}{2^k})+1+1)+1)+...+1)$
$ T(n) = \sum_{i=1}^{k}{1}$
$ T(n) = k$

Mas: $n=2^k$. Fazendo o $\log_{2}$ nos dois lados da equação:

$\log_{2}(n) = \log_{2}{2^k}$

$\log_{2}(n) = k\log_{2}{2}$

$\log_{2}(n) = k$.
Ou seja:

$ T(n) = \log_{2}(n)$

Assim, a complexidade do algoritmo é $O(\log_{2}(n))$.





Asssumir $T(1) =1 $ é a chamada condição de parada, sendo 1 o valor base da recorrência, ou seja, o valor para o qual a recorrência é interrompida

E para o caso em que temos:

$T(n) = T(n/2) + n$, com $T(1)=$?

Expandimos a relação:

$T(n) = T(n/2) + n$
$T(n) = (T(n/4)+ n/2) + n$
$T(n) = ((T(n/8) + n/4)+n/2)+n$
...
$T(n) = (...((T(n/2^k) + n/2^{k-1})+n/2^{k-2})+...+n/2)+n$

Com $n = 2^{k}$
$T(n) = T(1) + 2+ 4 +8+16+...+2^{k}$



A partir da expansão anterior, podemos ver que um bom valor para $T(1)$ é 1, pois isso facilita o fechamento do somatório. Assim:
$T(n) = 1 + 2+ 4 +8+16+...+2^{k}$

$T(n) = \sum_{i=0}^{k}2^{i}$

Isso é uma Progressão Geométrica de Ordem 2, com termo geral sendo $a_{k} = a_{1}q^{k-1}$.

A somatóra dos termos da PG é:

$S_{n} = \frac{a_{1}(q^{k}-1)}{q-1}$









Levando em conta a seguinte propriedade de somatório:

$\sum_{i=0}^{k}2^{i} = 2^{k}-1$

Lembrando que $n=2^{k}$ implica em $k = \log_{2}(n)$, ou seja:

$T(n) = \sum_{i=0}^{k}2^{i} = 2^{\log_{2}(n)}-1$

$T(n) = 2n-1$
